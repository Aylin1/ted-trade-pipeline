```python
import urllib.parse

def create_tradeatlas_link(
    hs_codes,
    firms,
    start_date="2018-01-01",
    end_date="2025-10-01",
    country="DE",
    max_hs=10,
    max_firms=10
):
    """
    Generate TradeAtlas shipment search URLs with encoded HS codes and firm names.
    
    Args:
        hs_codes (list): List of HS code strings.
        firms (list): List of firm names (will be simplified and truncated if long).
        start_date (str): Start date (YYYY-MM-DD).
        end_date (str): End date (YYYY-MM-DD).
        country (str): Importing country (e.g. 'DE').
        max_hs (int): Max number of HS codes per URL (default 10).
        max_firms (int): Max number of firms per URL (default 10).

    Returns:
        list: List of TradeAtlas URLs.
    """

    # Split lists into chunks
    def chunk_list(lst, n):
        for i in range(0, len(lst), n):
            yield lst[i:i + n]

    links = []
    base_url = "https://app.tradeatlas.com/en/shipment-search"

    for hs_chunk in chunk_list(hs_codes, max_hs):
        hs_params = ",".join([f"9166%3A{hs}" for hs in hs_chunk])

        for firm_chunk in chunk_list(firms, max_firms):
            # Use only the first word of each company name, exclude "GmbH" etc.
            simplified = [f.split()[0] for f in firm_chunk if f.strip()]
            firm_str = "%2C%257C".join([urllib.parse.quote(f) for f in simplified])

            url = (
                f"{base_url}?dates={start_date}%2C{end_date}"
                f"&hs={hs_params}"
                f"&icnty={country}"
                f"&ifirm={firm_str}"
                f"&firms=0"
            )
            links.append(url)

    return links
```

```python
hs_codes = [
    "610510", "610910", "611420", "621010", "621133", 
    "630231", "610610", "611300", "611510", "620630",
    "630260", "621142", "621143", "611610"
]

firms = [
    "Ziegler Textil GmbH", "Reiner Steinmann GmbH", "Bewernick GmbH Textilfabrik",
    "Famo Vertrieb", "Recon Company", "Lyreco Deutschland GmbH",
    "Detlef Paulsen Betriebsausrüster GmbH", "Schloemer GmbH", 
    "Wunsch & Rudolph e.K.", "Coenen Neuss GmbH", "A+S Technigro GmbH"
]

links = create_tradeatlas_link(hs_codes, firms)

for link in links:
    print(link)
```
